```python
import numpy as np
from sklearn.mixture import GMM
from scipy import linalg
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
# column titles: catid,coljk,colk3,col12,col23,col34,stype,o/c class)
# for the moment, we only want w1-w2, w2-w3, w3-w4, and o/c class
f1 = '../selected_known_agbs/SIMBAD_c_rich_testset.dat'
f2 = '../selected_known_agbs/SIMBAD_o_rich_testset.dat'

## ==========================
## Choose my dimensions here
## ==========================
data1 = np.loadtxt(f1, usecols = (3,4,5,7))
data2 = np.loadtxt(f2, usecols = (3,4,5,7))
```

```python
X1 = data1[:, :-1] ## C-rich
X2 = data2[:, :-1] ## O-rich

y1 = data1[:,-1]
y2 = data2[:,-1]

X = np.concatenate((X1,X2))
y = np.concatenate((y1,y2))

col12min = min(X[:,0])
col12max = max(X[:,0])

col23min = min(X[:,1])
col23max = max(X[:,1])

col34min = min(X[:,2])
col34max = max(X[:,2])

xlim = (col12min, col12max) 
ylim = (col23min, col23max)
zlim = (col34min, col34max)
```

```python
n_classes = len(np.unique(y))

clf = GMM(n_components = n_classes, covariance_type='full', 
          init_params="wc", n_iter=1)
clf.means_ = np.array([X[y == i].mean(axis=0) for i in xrange(n_classes)])
clf.fit(X)
```

```python
## Read in the AGB stars with the bulge relation

## each infile has the following columns:
##
## wid,ra,decl,gall,galb
## w1,w2,w3,w4,w1err,w2err,w3err,w4err,w1snr,w2snr,w3snr,w4snr
## pmra,pmdec,sigpmra,sigpmdec
## j,h,k,jerr,herr,kerr
## coljk,colk3,col12,col23,col34
## Mw1, E(B-V)

data_dir = '/Users/Nick/Documents/AGBstuff/workstation/allwise_reboot/allwise/candidates_condor/candidate_outs_bulge/'
infiles = [data_dir+f for f in os.listdir(data_dir) if f.endswith('.dat')]

n = len(infiles)

wcnt = []
ra = []
decl = []
gall = []
galb = []
w1 = []
w2 = []
w3 = []
w4 = []
w1err = []
w2err = []
w3err = []
w4err = []
w1snr = []
w2snr = []
w3snr = []
w4snr = []
pmra = []
pmdec = []
sigpmra = []
sigpmdec = []
j = []
h = []
k = []
jerr = []
herr = []
kerr = []
coljk = []
colk3 = []
col12 = []
col23 = []
col34 = []
absw1 = []
ebvs = []

for i in range(len(infiles)):
    data = np.loadtxt(infiles[i], unpack=True)
    if len(data) > 0:
        wcnt.append(data[0])
        ra.append(data[1])
        decl.append(data[2])
        gall.append(data[3])
        galb.append(data[4])
        w1.append(data[5])
        w2.append(data[6])
        w3.append(data[7])
        w4.append(data[8])
        w1err.append(data[9])
        w2err.append(data[10])
        w3err.append(data[11])
        w4err.append(data[12])
        w1snr.append(data[13])
        w2snr.append(data[14])
        w3snr.append(data[15])
        w4snr.append(data[16])
        pmra.append(data[17])
        pmdec.append(data[18])
        sigpmra.append(data[19])
        sigpmdec.append(data[20])
        j.append(data[21])
        h.append(data[22])
        k.append(data[23])
        jerr.append(data[24])
        herr.append(data[25])
        kerr.append(data[26])
        coljk.append(data[27])
        colk3.append(data[28])
        col12.append(data[29])
        col23.append(data[30])
        col34.append(data[31])
        absw1.append(data[32])
        ebvs.append(data[33])
    
wcnt = np.concatenate(wcnt)
ra = np.concatenate(ra)
decl = np.concatenate(decl)
gall = np.concatenate(gall)
galb = np.concatenate(galb)
w1 = np.concatenate(w1)
w2 = np.concatenate(w2)
w3 = np.concatenate(w3)
w4 = np.concatenate(w4)
w1err = np.concatenate(w1err)
w2err = np.concatenate(w2err)
w3err = np.concatenate(w3err)
w4err = np.concatenate(w4err)
w1snr = np.concatenate(w1snr)
w2snr = np.concatenate(w2snr)
w3snr = np.concatenate(w3snr)
w4snr = np.concatenate(w4snr)
pmra = np.concatenate(pmra)
pmdec = np.concatenate(pmdec)
sigpmra = np.concatenate(sigpmra)
sigpmdec = np.concatenate(sigpmdec)
j = np.concatenate(j)
h = np.concatenate(h)
k = np.concatenate(k)
jerr = np.concatenate(jerr)
herr = np.concatenate(herr)
kerr = np.concatenate(kerr)
coljk = np.concatenate(coljk)
colk3 = np.concatenate(colk3)
col12 = np.concatenate(col12)
col23 = np.concatenate(col23)
col34 = np.concatenate(col34)
absw1 = np.concatenate(absw1)
ebvs = np.concatenate(ebvs)
```

```python
coeff = [0.34,0.21,0.15,0.11,0.09,0.13,0.13]
Av = 3.1*ebvs
Ar = 2.751*ebvs
Aj = coeff[0]*2.751*ebvs
Ah = coeff[1]*2.751*ebvs
Ak = coeff[2]*2.751*ebvs
Aw1 = coeff[3]*2.751*ebvs
Aw2 = coeff[4]*2.751*ebvs
Aw3 = coeff[5]*2.751*ebvs
Aw4 = coeff[6]*2.751*ebvs
DM = w1 - absw1 - Aw1
dist = 10**((DM+5.)/5.)/1E3 
```
