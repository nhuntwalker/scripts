<b>Purpose:</b> Use the relationships between infrared colors and mass loss rates from Le Bertre 1997 and Le Bertre & Winters 1998 to derive mass loss rates for my AGB candidates. The formulae are divided by spectra class, so I'll have to reclassify my stars (since I stupidly didn't save that information).  The relationships are...

<b>O-rich Stars (from Le Bertre & Winters 1998)</b>
\begin{eqnarray}
\log_{10}\dot{M} &=& \displaystyle\frac{-2.5}{[J-K] - 0.65} + 1.75; 1.6 \le J-K \le 6.0\\
\log_{10}\dot{M} &=& \displaystyle\frac{-2.75}{[K-L]} + 2.25; 0.7 \le K-L \le 3.0\\
\log_{10}\dot{M} &=& \displaystyle\frac{-55}{[K-12] + 5} + 6; 2 \le K-12 \le 7\\
\end{eqnarray}

<b>C-rich Stars (from Le Bertre 1997 and Le Bertre & Winters 1998)</b>
\begin{eqnarray}
\log_{10}\dot{M} &=& \displaystyle\frac{-6.0}{[J-K] - 0.2} + 2.0; 2 \le J-K \le 7\\
\log_{10}\dot{M} &=& \displaystyle\frac{-9.0}{[K-L] + 1.4} + 2.75; 1 \le K-L \le 8\\
\log_{10}\dot{M} &=& \displaystyle\frac{-24}{[K-12] + 4} + 3; 2 \le K-12 \le 14\\
\end{eqnarray}
