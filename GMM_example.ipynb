```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.mlab as mlab
import numpy as np
from sklearn import mixture
from scipy import linalg
import itertools

%matplotlib inline
```

```python
loc = 10
n = 1000
x1 = np.random.normal(size=n)
x2 = np.random.normal(loc=loc,size=n)
a = np.concatenate((x1,x2))

y1 = np.random.normal(size=n)
y2 = np.random.normal(loc=loc,size=n)
b = np.concatenate((y1,y2))

X_train = np.vstack((a,b)).T

plt.figure(figsize=(5,5))
plt.scatter(a,b,s=1)
plt.ylim(-20,20)
plt.xlim(-20,20)
```

```python
clf = mixture.GMM(n_components=2, covariance_type = 'full')
clf.fit(X_train)

dpgmm = mixture.DPGMM(n_components=2, covariance_type = 'full')
dpgmm.fit(X_train)
```

```python
x = np.linspace(-20.,20.)
y = np.linspace(-20.,20.)
X1, Y1 = np.meshgrid(x,y)
XX1 = np.c_[X1.ravel(), Y1.ravel()]
Z1 = np.log(-clf.score_samples(XX1)[0])
Z1 = Z1.reshape(X1.shape)

x = np.linspace(-5.,15.,100)
y = np.linspace(-5.,15.,100)
X2, Y2 = np.meshgrid(x,y)
XX2 = np.c_[X2.ravel(), Y2.ravel()]
Z2 = np.exp(dpgmm.score_samples(XX2)[0])
Z2 = Z2.reshape(X2.shape)
```

```python
Z2.shape
```

```python
color_iter= itertools.cycle(['r','b'])
plt.figure(figsize=(12,5))
ax = plt.subplot(121)
ax.scatter(a,b,s=1)

for i, (mean, covar, color) in enumerate(zip(
     clf.means_, clf._get_covars(), color_iter)):
    v, w = linalg.eigh(covar)
    u = w[0]/linalg.norm(w[0])
    
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi
    ell = Ellipse(mean, v[0], v[1], 180 + angle, color=color)
    ell.set_clip_box(ax.bbox)
    ell.set_alpha(0.5)
    ax.add_artist(ell)

CS = ax.contour(X1, Y1, Z1)
CB = plt.colorbar(CS, shrink=0.8, extend='both')
ax.set_ylim(-5,15)
ax.set_xlim(-5,15)

ax = plt.subplot(122)
ax.scatter(a,b,s=1)

for i, (mean, covar, color) in enumerate(zip(
     dpgmm.means_, dpgmm._get_covars(), color_iter)):
    v, w = linalg.eigh(covar)
    u = w[0]/linalg.norm(w[0])
    
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi
    ell = Ellipse(mean, v[0], v[1], 180 + angle, color=color)
    ell.set_clip_box(ax.bbox)
    ell.set_alpha(0.5)
    ax.add_artist(ell)

CS = ax.contour(X2, Y2, Z2)
CB = plt.colorbar(CS, shrink=0.8, extend='both')
ax.set_ylim(-5,15)
ax.set_xlim(-5,15)
```

```python
means = clf.means_.T[0]
xsigs = [clf._get_covars()[0,0,0],clf._get_covars()[1,0,0]]

xarr = np.linspace(-5,20,101)
plt.hist(X_train.T[1], bins=np.linspace(-5,15,101))
plt.plot(xarr, 150*mlab.normpdf(xarr, means[0], xsigs[0]), color='r')
plt.plot(xarr, 150*mlab.normpdf(xarr, means[1], xsigs[1]), color='y')
```
