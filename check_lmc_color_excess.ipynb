```python
gall = np.linspace(274.,286.,50)
galb = np.linspace(-38.,-26.,50)

coords = []
for ii in range(len(gall)):
    for jj in range(len(galb)):
        coords.append([float(gall[ii]),float(galb[jj])])
        if (ii==1) & (jj==1):
            print gall[ii],galb[ii]

glons,glats = np.array(coords).T
```

```python
## Convert gall/galb to radians
conv = np.pi/180.
glons = glons*conv
glats = glats*conv
```

```python
## Use a distance modulus of 18.32 to create an array of distance moduli, 
## then run galfastDust

import galfastDust as ext

DMarry = np.zeros(len(glons)) + 18.32
dist_arry = 10**((DMarry+5.)/5.)/1E3 #convert DM to kpc
ebvs_lmc = np.array(ext.getDustValue(glats,glons,dist_arry))
```

```python
Aw1_lmc = 0.11*2.751*ebvs_lmc
gall_lmc = glons/conv
galb_lmc = glats/conv
H2, xed, yed = binned_statistic_2d(gall_lmc, galb_lmc, Aw1_lmc, bins=50)
```

```python
extent = [xed[0],xed[-1],yed[0],yed[-1]]
fig = plt.figure(figsize=(8,8))
ax = plt.subplot(111)
plt.title('2D Histogram of A$_{W1}$\nfrom the SFD Dust Maps')
im = plt.imshow(H2.T, extent=extent, aspect='auto', interpolation='nearest', cmap=plt.cm.Greys,vmin=0, vmax=3)

plt.ylabel('glat (deg)')
plt.xlabel('glon (deg)')
plt.minorticks_on()

cax = fig.add_axes([0.92,0.125,0.03,0.775])
cbar = fig.colorbar(im, cax=cax)
cbar.ax.set_ylabel('A$_{W1}$')

ax.set_xlim(274,286)
ax.set_ylim(-38,-26)
plt.show()
```
