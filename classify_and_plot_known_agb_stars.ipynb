We will fit our Gaussian Mixture Models to 3 attributes

```python
X1 = data1[:, :-1] ## C-rich
X2 = data2[:, :-1] ## O-rich

y1 = data1[:,-1]
y2 = data2[:,-1]

X = np.concatenate((X1,X2))
y = np.concatenate((y1,y2))

col12min = min(X[:,0])
col12max = max(X[:,0])

col23min = min(X[:,1])
col23max = max(X[:,1])

col34min = min(X[:,2])
col34max = max(X[:,2])

xlim = (col12min, col12max) 
ylim = (col23min, col23max)
zlim = (col34min, col34max)
```

```python
n_classes = len(np.unique(y))

clf = GMM(n_components = n_classes, covariance_type='full', 
          init_params="wc", n_iter=1)
```

```python
clf.means_ = np.array([X[y == i].mean(axis=0) for i in xrange(n_classes)])
clf.fit(X)
```

```python
def make_ellipses123(gmm, ax, n_classes):
    ## Ellipses for w1-w2 vs w2-w3
    for n, color in enumerate('br'):
        v, w = np.linalg.eigh(gmm._get_covars()[n][:2,:2])
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi
        v *= 9
        ell = Ellipse(gmm.means_[n, :2], v[0], v[1],
                      180 + angle, color=color, linewidth=2)
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        ax.add_artist(ell)

def make_ellipses234(gmm, ax, n_classes):
    ## Ellipses for w2-w3 vs w3-w4
    for n, color in enumerate('br'):
        v, w = np.linalg.eigh(gmm._get_covars()[n][1:,1:])
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi
        v *= 9
        ell = Ellipse(gmm.means_[n, 1:], v[0], v[1],
                      180 + angle, color=color, linewidth=2)
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        ax.add_artist(ell)
```

Because we don't yet know the intrinsic distances to each of these stars, we will need to calculate E(B-V) values for a set of distances in the direction of each star. So, we have a loop for each star.

1) Select star and open "for" loop iterating over distance array D

2) For each distance in the direction of ra/dec, calculate E(B-V)

3) $m_\lambda = M_\lambda + (5 \log(D[$pc$]) - 5) + A_\lambda$

4) We know $m_\lambda$ from observations, and from the color-magnitude relations for $M_\text{W1}$ we can find the absolute magnitudes from the star's color. We want to solve for $A_\lambda$ by numerically solving for $DM = (5 \log(D[$pc$]) - 5)$ and $A_\lambda$ at the same time.

5) $\Sigma = m_\lambda - M_\lambda$ = $DM + A_\lambda$

6) $A_\lambda(DM) = \Sigma - DM$

7) We adopt a $DM$ and $A_\lambda$ when $A_\lambda$ calculated from the extinction maps matches the $A_\lambda$ from the above equation.

8) Note that we don't have to iterate over the same distance array for each.  The maximum distance for any star will be $\Sigma$. So, we can go from zero to $\Sigma$ in even steps.

```python
## We define a function that can calculate the absolute magnitude for 
## a given star based on its colors and its spectral class.
##
## Note: these values are based on color magnitude relations for
## AGB stars in the direction of the LMC only.

def spectype_relation(colk3,col12,col23,y):
    absw1 = np.zeros((len(y),3))
    orich = np.array([[-9.15,0.02],[-9.13,0.05],[-9.00,0.09]])
    crich = np.array([[-9.16,0.03],[-9.46,0.42],[-9.33,0.18]])

    xvars = np.array([colk3,col12,col23])
    for ii in range(3):
        absw1[y == 0,ii] = orich[ii,0] + orich[ii,1]*xvars[ii,y==0]## w1 vs colk3
        absw1[y == 1,ii] = crich[ii,0] + crich[ii,1]*xvars[ii,y==1]## w1 vs colk3

    return absw1
```

```python
def regress(xvals,yvals,deg=1):
    ## For n-degree regression
    ordered = sorted(range(len(xvals)), key=lambda k: xvals[k])
    X = xvals[ordered]
    
    Mtuple = []
    Mtuple.append(np.ones((len(X))))
    
    for ii in range(deg):
        Mtuple.append(X**(ii+1))

    M = np.column_stack(Mtuple)
    
    Y = yvals[ordered]
    
    A = np.dot(M.transpose(),M)
    B = np.dot(M.transpose(),Y)
    
    theta = np.dot(np.linalg.pinv(A),B)
    
    Y_model = theta[0]
    for ii in range(deg):
        Y_model += theta[ii+1]*X**(ii+1)
    
    return X, Y_model, theta
```
