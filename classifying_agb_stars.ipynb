```python
from sklearn import mixture
from scipy import linalg
from matplotlib.patches import Ellipse
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os

os.chdir('/Users/Nick/Documents/AGBstuff/workstation/allwise_reboot/scripts')
```

```python
# column titles: catid,coljk,colk3,col12,col23,col34,stype,o/c class)
# for the moment, we only want color23, 34, and o/c class

f1 = '../selected_known_agbs/SIMBAD_c_rich_testset.dat'
f2 = '../selected_known_agbs/SIMBAD_o_rich_testset.dat'

## ==========================
## Choose my dimensions here
## ==========================
data1 = np.loadtxt(f1, usecols = (4,5,7))
data2 = np.loadtxt(f2, usecols = (4,5,7))
```

```python
## ================
## Attributes here
## ================
X1 = data1[:, :-1] ## C-rich
X2 = data2[:, :-1]

## ==================
## Point labels here
## ==================
y1 = data1[:,-1]
y2 = data2[:,-1]

## ==================================
## Setting my graph limits for later
## ==================================
X = np.concatenate((X1,X2))
y = np.concatenate((y1,y2))


col23min = min(X[:,0])
col23max = max(X[:,0])

col34min = min(X[:,1])
col34max = max(X[:,1])

xlim = (col23min, col23max) 
ylim = (col34min, col34max)
```

```python
## =============================================================
## Return regularly-spaced coordinate matrices from 
## 2 or more coordinate vectors.
##
## Note that you need to make an x-y grid for every two
## attribute combinations that you wish to view/fit to
## =============================================================
xgrid, ygrid = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                           np.linspace(ylim[0], ylim[1], 50))

## ===============================================================
## This takes xgrid (an array of regularly-spaced x points for
## every y-point) and ygrid (vice versa) and produces coordinates 
## for every point in your desired grid.
## Note that this is only if you want to visualize your Gaussians
## as a "score" on a grid.
## ===============================================================
datagrid = np.c_[xgrid.ravel(), ygrid.ravel()]

## =============================
## Select GMMs as my classifier
## =============================
clf1 = mixture.GMM(n_components=1, n_iter=10, covariance_type = 'full')
clf2 = mixture.GMM(n_components=1, n_iter=10, covariance_type = 'full')

## =================================================
## Fit my data here.  This is literally the easiest
## part of this whole process.
##
## Note that this is independent of the grid we set
## up earlier. When we visualize, the fit will be
## set to whatever grid we want
## =================================================
clf1.fit(X1)
clf2.fit(X2)
```

```python
## ================================================================
## Want to do a coordinate transform to plot the ellipses on their
## respective angles
## ================================================================
## C-rich Ellipse
covar = clf1._get_covars()[0]
v, w = linalg.eigh(covar)
#v = covar.diagonal()
u = w[0] / linalg.norm(w[0])
angle = np.arctan(u[1] / u[0])
angle = 180*angle/np.pi # convert to degrees
ell1 = Ellipse(clf1.means_[0], 20*v[0], 20*v[1], 180+angle, color='r')

## O-rich Ellipse
covar = clf2._get_covars()[0]
v, w = linalg.eigh(covar)
#v = covar.diagonal()
u = w[0] / linalg.norm(w[0])
angle = np.arctan(u[1] / u[0])
angle = 180*angle/np.pi # convert to degrees
ell2 = Ellipse(clf2.means_[0], 20*v[0], 20*v[1], 180+angle, color='b')
```

```python
## ================================
## Let's plot and see what we get!
## ================================
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')
ax.scatter(X1[:, 0], X1[:, 1], c='k', s=3, edgecolor='None') #crich
ax.scatter(X2[:, 0], X2[:, 1], c='g', s=3, edgecolor='None') #orich

ell1.set_clip_box(ax.bbox)
ell1.set_alpha(0.5) ## to make semi-transparent
ax.add_artist(ell1)

ell2.set_clip_box(ax.bbox)
ell2.set_alpha(0.5) ## to make semi-transparent
ax.add_artist(ell2)

ax.set_xlim(xlim)
ax.set_xlabel('W2-W3')
ax.set_ylim(ylim)
ax.set_ylabel('W3-W4')

plt.show()
```
