```python
font = {'family':'normal', 'weight':'bold', 'size':10}

def sixplots(gall=gall,galb=galb,w1=w1,w2=w2,w3=w3,w4=w4,j=j,h=h,k=k,font=font,cut=None,title=None,scatter='no',savename=None):
	if cut != None:
		gall = gall[cut]
		galb = galb[cut]
		w1 = w1[cut]
		w2 = w2[cut]
		w3 = w3[cut]
		w4 = w4[cut]
		j = j[cut]
		h = h[cut]
		k = k[cut]

	col12 = w1-w2
	col23 = w2-w3
	col34 = w3-w4
	coljh = j-h
	colhk = h-k
	colk1 = k-w1
	coljk = j-k	
	colk3 = k-w3

	thick = 2
	tint = 0.3

	matplotlib.rc('font', **font)

	fig = plt.figure(figsize=(12,8.5))
	if title != None:
		fig.subplots_adjust(top=0.9,bottom=0.1,left=0.07,right=0.975, hspace=0.2)
		plt.suptitle(title)

	else:
		fig.subplots_adjust(top=0.975,bottom=0.1,left=0.07,right=0.975, hspace=0.2)
	## K vs J-K
	ax = plt.subplot(2,3,1)
	xbins = np.arange(1,10.1,0.1)
	ybins = np.arange(2,15.6,0.1)
	if scatter.lower() != 'no':
		ax.scatter(coljk,k,s=1,edgecolor='none',c='k')
	else:
		H, xed, yed = np.histogram2d(coljk,k,bins=(xbins,ybins))
		extent = [xed[0],xed[-1],yed[0],yed[-1]]
		ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
	ax.minorticks_on()
	ax.set_xlim(1,8)
	ax.set_ylim(15.5,2)
	ax.set_xlabel('J-K (mag)')
	ax.set_ylabel('K (mag)')

	## K-W3 vs J-K
	ax = plt.subplot(2,3,2)
	xbins = np.arange(1,10.1,0.1)
	ybins = np.arange(-2,10.5,0.1)
	if scatter.lower() != 'no':
		ax.scatter(coljk,colk3,s=1,edgecolor='none',c='k')
	else:
		H, xed, yed = np.histogram2d(coljk,colk3,bins=(xbins,ybins))
		extent = [xed[0],xed[-1],yed[0],yed[-1]]
		ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
	ax.minorticks_on()
	ax.set_xlim(1,8)
	ax.set_ylim(0,6)
	ax.set_xlabel('J-K (mag)')
	ax.set_ylabel('K-W3 (mag)')

	## K vs W2-W3
	ax = plt.subplot(2,3,3)
	xbins = np.arange(0,4.6,0.05)
	ybins = np.arange(2,15.6,0.1)
	if scatter.lower() != 'no':
		ax.scatter(col23,k,s=1,edgecolor='none',c='k')
	else:
		H, xed, yed = np.histogram2d(col23,k,bins=(xbins,ybins))
		extent = [xed[0],xed[-1],yed[0],yed[-1]]
		ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
	ax.minorticks_on()
	ax.set_xlim(0,3.5)
	ax.set_ylim(15.5, 2)
	ax.set_xlabel('W2-W3 (mag)')
	ax.set_ylabel('K (mag)')

	## K-W3 vs W1-W2
	ax = plt.subplot(2,3,4)
	xbins = np.arange(-1.5,4.6,0.05)
	ybins = np.arange(-2,10.5,0.1)
	if scatter.lower() != 'no':
		ax.scatter(col12,colk3,s=1,edgecolor='none',c='k')
	else:
		H, xed, yed = np.histogram2d(col12,colk3,bins=(xbins,ybins))
		extent = [xed[0],xed[-1],yed[0],yed[-1]]
		ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
	ax.minorticks_on()
	ax.set_xlim(-1.,2.)
	ax.set_ylim(0,6)
	ax.set_xlabel('W1-W2 (mag)')
	ax.set_ylabel('K-W3 (mag)')

	## W1-W2 vs W2-W3
	ax = plt.subplot(2,3,5)
	xbins = np.arange(-1.5,4.6,0.05)
	ybins = np.arange(0,4.6,0.05)
	if scatter.lower() != 'no':
		ax.scatter(col23,col12,s=1,edgecolor='none',c='k')
	else:
		H, xed, yed = np.histogram2d(col23,col12,bins=(xbins,ybins))
		extent = [xed[0],xed[-1],yed[0],yed[-1]]
		ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)

	ax.fill_between([0,4], [0.629*(0) - 0.198,0.629*4 - 0.198], [0.629*(0) + 0.359,0.629*4 + 0.359], color='r', alpha=tint)
	ax.plot([0,4], [0.629*(0) + 0.359,0.629*4 + 0.359], color='r')
	ax.plot([0,4], [0.629*(0) - 0.198,0.629*4 - 0.198], color='r')
	ax.plot([0,0], [-1,4], color='k')
	ax.plot([4,4], [-1,4], color='k')

	ax.fill_between([0.2,2.],[0.4,0.4],[2.2,2.2], color='b', alpha=tint)
	ax.plot([0.2,0.2], [0.4,2.2], color='b')
	ax.plot([2.0,2.0], [0.4,2.2], color='b')

	ax.minorticks_on()
	ax.set_xlim(0,4.)
	ax.set_ylim(0,4.)
	ax.set_xlabel('W1-W2 (mag)')
	ax.set_ylabel('W2-W3 (mag)')

	## W3-W4 vs W2-W3
	ax = plt.subplot(2,3,6)
	xbins = np.arange(0,4.6,0.05)
	ybins = np.arange(-1.5,4.6,0.05)
	if scatter.lower() != 'no':
		ax.scatter(col23,col34,s=1,edgecolor='none',c='k')
	else:
		H, xed, yed = np.histogram2d(col23,col34,bins=(xbins,ybins))
		extent = [xed[0],xed[-1],yed[0],yed[-1]]
		ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)

	ax.fill_between([0,4], [0.722*(0) - 0.911,0.722*4 - 0.911], [0.722*(0) - 0.289,0.722*4 - 0.289], color='r', alpha=tint)
	ax.plot([0,4], [0.722*(0) - 0.289,0.722*4 - 0.289], color='r')
	ax.plot([0,4], [0.722*(0) - 0.911,0.722*4 - 0.911], color='r')
	ax.plot([0,0], [-1,4], color='k')
	ax.plot([4,4], [-1,4], color='k')

	ax.fill_between([0.4,2.2],[0.722*0.4-0.289, 0.722*2.2-0.289],[1.3,1.3], color='b', alpha=tint)
	ax.plot([0.4,2.2], [1.3,1.3], color='b')
	ax.plot([0.4,0.4], [0,1.3], color='b')
	ax.plot([0.4,2.2], [0.722*0.4-0.289, 0.722*2.2-0.289], color='b')
	ax.plot([0,0], [-1,4], color='k')
	ax.plot([4,4], [-1,4], color='k')

	ax.minorticks_on()
	ax.set_xlim(0,3.5)
	ax.set_ylim(-1.5,3.5)
	ax.set_xlabel('W2-W3 (mag)')
	ax.set_ylabel('W3-W4 (mag)')

	if savename != None:
		plt.savefig(savename)

	plt.show()

lmc_ra = 80.8917
lmc_dec = -69.7561

smc_ra = 13.187
smc_dec = -72.829

deglmc = 15.
degsmc = 10.

radlmc = ((ra - lmc_ra)/3.5)**2 + ((decl - lmc_dec)/1.5)**2
radlmc_z = ((ra_z - lmc_ra)/3.5)**2 + ((decl_z - lmc_dec)/1.5)**2
radsmc = ((ra - smc_ra)/3.5)**2 + ((decl - smc_dec)/1.5)**2
radsmc_z = ((ra_z - smc_ra)/3.5)**2 + ((decl_z - smc_dec)/1.5)**2

cutlmc = np.where(radlmc < deglmc)
cutlmc_z = np.where(radlmc_z < deglmc)
cutsmc = np.where(radsmc < degsmc)
cutsmc_z = np.where(radsmc_z < degsmc)

cut1 = np.where(abs(galb) > 3)
cut2 = np.where((abs(galb) < 3) & (abs(gall) < 45))
cut3 = np.where((abs(galb) < 3) & (abs(gall) > 45))

title_lmc = 'LMC population\nZeljko Allsky'
title_smc = 'SMC population\nZeljko Allsky'
title1 = '|glat| > 3\nZeljko Allsky'
title2 = '|glat| < 3 and |glon| < 45\nZeljko Allsky'
title3 = '|glat| < 3 and |glon| > 45\nZeljko Allsky'

sixplots(gall,galb,w1_z,w2_z,w3_z,w4_z,j_z,h_z,k_z, cut=cut1,title=title1)
sixplots(gall,galb,w1_z,w2_z,w3_z,w4_z,j_z,h_z,k_z, cut=cut2,title=title2)
sixplots(gall,galb,w1_z,w2_z,w3_z,w4_z,j_z,h_z,k_z, cut=cut3,title=title3)
sixplots(gall,galb,w1_z,w2_z,w3_z,w4_z,j_z,h_z,k_z, cut=cutlmc,title=title_lmc, scatter="yes")
sixplots(gall,galb,w1_z,w2_z,w3_z,w4_z,j_z,h_z,k_z, cut=cutsmc,title=title_smc, scatter="yes")

# sixplots(cut=cut1,title=title1)
# sixplots(cut=cut2,title=title2)
# sixplots(cut=cut3,title=title3)
# sixplots(cut=cutlmc,title=title_lmc,scatter='yes')
# sixplots(cut=cutsmc,title=title_smc,scatter='yes')
```
