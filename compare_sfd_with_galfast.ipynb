Here I'm comparing the A$_{W1}$ derived from GalFast to the value derived from the SFD dust maps. The last cell has a plot that shows the comparison between the full dust columns in SFD maps, and the dust columns from GalFast at 8 kpc. The latitude range is |$gal\;\mathcal{l}$| $<$ 10. The longitude range is |$gal\;\mathcal{b}$| $<$  10.



```python
import numpy as np
import matplotlib.pyplot as plt
import os
from astroML.stats import binned_statistic_2d
%matplotlib inline
```

```python
os.chdir('/Users/Nick/Documents/AGBstuff/workstation/allwise_reboot/bulgetest')
infile1 = 'bulge_ebvs_galfast.dat'
infile2 = 'idlouts/all_ebvs.dat'
```

```python
gall_gf, galb_gf, ebvs_gf = np.loadtxt(infile1, unpack=True)
gall_sfd, galb_sfd, ebvs_sfd = np.loadtxt(infile2, unpack=True)
# data = open(infile1).readlines()

Aw1_gf = 0.11*2.751*ebvs_gf
Aw1_sfd = 0.11*2.751*ebvs_sfd
```

```python
gall_sfd[gall_sfd > 180] = gall_sfd[gall_sfd > 180] - 360
gall_gf[gall_gf > 180] = gall_gf[gall_gf > 180] - 360
```

```python
xbins = np.linspace(-10,10,151)
ybins = np.linspace(-10,10,151)
```

```python
H1, xed, yed = binned_statistic_2d(gall_sfd, galb_sfd, Aw1_sfd, bins=(xbins,ybins))
H2, xed, yed = binned_statistic_2d(gall_gf, galb_gf, Aw1_gf, bins=(xbins,ybins))
```

```python
extent = [xed[0],xed[-1],yed[0],yed[-1]]
fig = plt.figure(figsize=(8,8))
fig.subplots_adjust(top=1.0, hspace=0)
ax = plt.subplot(311)
# im = ax.imshow(H.T, extent=extent, aspect='auto', interpolation='nearest', cmap=plt.cm.Greys)
im = ax.imshow(H1.T, extent=extent, aspect='auto', interpolation='nearest', cmap=plt.cm.Greys, vmax=10)
ax.text(0.9,0.9,'From\nSFD',horizontalalignment='right',verticalalignment='top',transform=ax.transAxes)
ax.set_ylabel('glat (deg)')
ax.set_xlabel('glon (deg)')
ax.set_xlim(-9.9,9.9)
ax.set_ylim(-9.9,9.9)
ax.minorticks_on()

cax = fig.add_axes([0.92,0.5,0.03,0.45])
cbar = fig.colorbar(im, cax=cax)
cbar.ax.set_ylabel('Logarithmic A$_{W1}$')

ax = plt.subplot(312)
# im = ax.imshow(H.T, extent=extent, aspect='auto', interpolation='nearest', cmap=plt.cm.Greys)
im = ax.imshow(H2.T, extent=extent, aspect='auto', interpolation='nearest', cmap=plt.cm.Greys, vmax=10)
ax.text(0.9,0.9,'From\nGalFast\nDistance = 8 kpc',horizontalalignment='right',verticalalignment='top',transform=ax.transAxes)
ax.set_ylabel('glat (deg)')
ax.set_xlabel('glon (deg)')
ax.set_xlim(-9.9,9.9)
ax.set_ylim(-9.9,9.9)
ax.minorticks_on()

ax = plt.subplot(313)
im = ax.imshow(H2.T-H1.T, extent=extent, aspect='auto', interpolation='nearest', vmax=2)
ax.text(0.9,0.9,'GalFast/SFD',horizontalalignment='right',verticalalignment='top',transform=ax.transAxes)
ax.set_ylabel('glat (deg)')
ax.set_xlabel('glon (deg)')
ax.set_xlim(-9.9,9.9)
ax.set_ylim(-9.9,9.9)
ax.minorticks_on()

cax = fig.add_axes([0.92,0.175,0.03,0.2])
cbar = fig.colorbar(im, cax=cax)
cbar.ax.set_ylabel('GalFast A$_{W1}$/SFD A$_{W1}$')

plt.show()
```
