Want to model the galaxy's gas distribution. Let's define some constants.

```python
#parameters of the radial density functions for HI and H2
a = [7.0,1.2] #kpc
b = [1.9,3.5] #kpc
c = [0.7,58.0] #cm^-3
```

```python
def radial_density(a,b,c,dist, H2=False):
    """Make sure dist is not iterable"""
#     if dist >= 1.2: kpc
    term1 = dist/a
    term2 = (b/dist)**2
    nH = c*np.exp(-term1 - term2)
    
    if (dist < 1.2) & (H2 == True):
        ## For H2 gas only
        alph = 0.1 #kpc
        beta = 240. #cm^-3
        nH = beta*np.exp(-(dist/alph)**2)
    
    return nH
```

```python
rad = np.linspace(0, 20., 100)
H1gas = [radial_density(a[0],b[0],c[0],d) for d in rad]
H2gas = [radial_density(a[1],b[1],c[1],d,H2=True) for d in rad]

plt.plot(rad, H1gas, 'b', linewidth=2)
plt.plot(rad, H2gas, 'r', linewidth=1)
plt.ylim(0,1.5)
plt.xlabel('Galactocentric Radius [kpc]')
plt.ylabel('Gas Density [cm$^{-3}$')
plt.minorticks_on()
plt.show()
```

```python
def vertical_density(a, b, c, rad, z, H2=False):
    term1 = rad/a
    term2 = (b/rad)**2
    nH_r = c*np.exp(-term1 - term2)
    
    if (rad < 1.2) & (H2 == True):
        ## For H2 gas only
        alph = 0.1 #kpc
        beta = 240. #cm^-3
        nH_r = beta*np.exp(-(rad/alph)**2)

    z_half = 45.*1.8*np.exp(0.1*rad) #pc
    z_c = -20. #pc
    
    nH = nH_r*np.exp(((z-z_c)/(1.2*z_half))**2)
    return nH
```

```python
## Integrate!

def trap_integrator(func, interval):
    h = 0.0001
    x = np.arange(interval[0],interval[1]+h,h)

    result = 0
    for ii in range(len(x)-1):
        pt1 = h*func(x[ii])
        pt2 = h*(func(x[ii+1]) - func(x[ii]))/2.
        result += pt1 + pt2
    
    return result
```

```python
def trap_int2d(func, interval):
    dx = 0.01
    dy = 0.01
    x = np.arange(interval[0][0], interval[0][1]+dx, dx)
    y = np.arange(interval[1][0], interval[1][1]+dy, dy)
    
    result = 0
    
    for ii in range(len(x)-1):
        for jj in range(len(y)-1):
            xpt1 = dx*dy*func(x[ii],y[jj])
            xpt2 = dx*dy*(func(x[ii+1],y[jj+1]) - func(x[ii],y[jj]))/2.
#             print "x1: %f, x2: %f" % (xpt1, xpt2)
            result += xpt1 + xpt2
    
    return result


def trap_int2d_vec(func, interval):
    ## This is a vectorized version of the trapezoidal integrator
    dx = 0.001
    dy = 0.001
    x = np.arange(interval[0][0], interval[0][1]+dx, dx)
    y = np.arange(interval[1][0], interval[1][1]+dy, dy)
    
    X,Y = np.meshgrid(x,y)
    
    xpt1 = dx*dy*(func(X[:-1,:-1],Y[:-1,:-1]) + (func(X[1:,1:],Y[1:,1:]) - func(X[:-1,:-1],Y[:-1,:-1]))/2.)
    result = np.sum(xpt1)

    return result


def func1(x,y):
    return np.exp(x*y**2 + x)
```
