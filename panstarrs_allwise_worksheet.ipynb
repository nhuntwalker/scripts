<b>Purpose:</b> The purpose of this iPython notebook is to do some visualization and subsequent analysis of the matched data between PANSTARRS unresolved sources and cleaned sources in ALLWISE from within rough photometric limits. The ALLWISE selection criteria are as follows:
<ul>
<li>$J-K > 1.0$</li>
<li>$W2-W3 > 0.0$</li>
<li>$W3-W4 < -0.83(W2-W3) + 3.37$</li>
<li>$W1 < 16.83$</li>
<li>$W2 < 15.6$</li>
<li>$W3 < 11.32$</li>
<li>$W4 < 9.0$</li>
</ul>

Further, after ingestion into the database <b>astroDS</b>, sources were retrieved which were devoid of confusion/contamination flags in every band, weren't identified as extended sources by WISE, and possessed only 1 match to the 2MASS survey.

```python
## Time to define a bunch of lovely little functions

def plot_glatglon(gall,galb,histo=True,cmap=plt.cm.Greys):
    fig = plt.figure(figsize=(10,5))
    ax = plt.subplot(111)
    if histo==False:
        ax.scatter(gall,galb,s=1,edgecolor='None',c='k')
    else:
        xbins = np.linspace(-180,180,361)
        ybins = np.linspace(-90,90,181)
        H, xed, yed = np.histogram2d(gall,galb,bins=(xbins,ybins))
        extent=[xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', cmap=cmap)
    ax.set_xlim(-180,180)
    ax.set_ylim(-90,90)
    ax.set_xlabel('Galactic Longitude (deg)')
    ax.set_ylabel('Galactic Latitude (deg)')
    ax.minorticks_on()
    plt.show()

def plot_photometry(xdata,ydata,xlabels,ylabels,nplots=1,histo=True,contours=False,yaxismag=True,nrows=False,ncols=False,cmap=plt.cm.Greys):
    if nplots == 1:
        fig = plt.figure(figsize=(4,4))
        ax = plt.subplot(111)
        xspan = sp.mquantiles(xdata,[0.001,0.999])
        yspan = sp.mquantiles(ydata,[0.001,0.999])

        if histo == False:
            ax.scatter(xdata,ydata,s=1,edgecolor='None',c='k')
        else:
            xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
            ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
            H, xed, yed = np.histogram2d(xdata,ydata,bins=(xbins,ybins))
            extent=[xed[0],xed[-1],yed[0],yed[-1]]
            ax.imshow(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', cmap=cmap)
        
        if contours != False:
            xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
            ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
            H, xed, yed = np.histogram2d(xdata,ydata,bins=(xbins,ybins))
            allquants = np.log10(H.ravel())
            #levels = np.linspace(0,max(allquants),7)[1:-1]
            levels = [0.5,1.0,1.5,2.0,2.5]
            extent=[xed[0],xed[-1],yed[0],yed[-1]]
            im = ax.contour(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', colors='r',levels=levels)            
            
            cax = fig.add_axes([0.925, 0.125, 0.05, 0.775]) 
            cbar = plt.colorbar(im, cax = cax)  
            cbar.ax.set_ylabel('log$_{10}$(N)')

        ax.set_xlim(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1)
        if yaxismag == True:
            ax.set_ylim(np.round(yspan[1],1) + 0.1,np.round(yspan[0],1))
        else:
            ax.set_ylim(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1)
        ax.set_xlabel(xlabels)
        ax.set_ylabel(ylabels)
        ax.minorticks_on()
    
    if nplots > 1:
        if len(xdata) != nplots or len(ydata) != nplots:
            return "Can't make plots. The number of plots does not equal the number of inputs."
        else:
            if ncols == False:
                if nplots <= 3:
                    ncols = nplots
                else:
                    ncols = 3
            if nrows == False:        
                nrows = m.ceil(nplots/3.)
            
            fig = plt.figure(figsize=(4.4*ncols,4.4*nrows))
            fig.subplots_adjust(hspace=0.2, wspace=0.2)

            for ii in range(nplots):
                ax = plt.subplot(nrows,ncols,ii+1)
                xspan = sp.mquantiles(xdata[ii],[0.001,0.999])
                yspan = sp.mquantiles(ydata[ii],[0.001,0.999])
        
                if histo == False:
                    ax.scatter(xdata[ii],ydata[ii],s=1,edgecolor='None',c='k')
                else:
                    xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
                    ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
                    H, xed, yed = np.histogram2d(xdata[ii],ydata[ii],bins=(xbins,ybins))
                    extent=[xed[0],xed[-1],yed[0],yed[-1]]
                    ax.imshow(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', cmap=cmap)
                
                if contours != False:
                    xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
                    ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
                    H, xed, yed = np.histogram2d(xdata[ii],ydata[ii],bins=(xbins,ybins))
                    allquants = np.log10(H.ravel())
                    #levels = np.linspace(0,max(allquants),7)[1:-1]
                    levels = [0.5,1.0,1.5,2.0,2.5]
                    extent=[xed[0],xed[-1],yed[0],yed[-1]]
                    im = ax.contour(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', colors='r',levels=levels)            
                    
                    cax = fig.add_axes([0.925, 0.125, 0.05/ncols, 0.775]) 
                    cbar = plt.colorbar(im, cax = cax)  
                    cbar.ax.set_ylabel('log$_{10}$(N)')
        
                ax.set_xlim(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1)
                if yaxismag == True:
                    ax.set_ylim(np.round(yspan[1],1) + 0.1,np.round(yspan[0],1))
                else:
                    ax.set_ylim(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1)
                ax.set_xlabel(xlabels[ii])
                ax.set_ylabel(ylabels[ii])
                ax.minorticks_on()
            
    plt.show()

def plot_withdavenportlocus(xcolors, ycolors, xlocusname, ylocusname, xcolor_name, ycolor_name, histo=True, nplots=1,stellar_locus=stellar_locus):

    if nplots == 1:
        xspan = sp.mquantiles(xcolors,[0.001,0.999])
        yspan = sp.mquantiles(ycolors,[0.001,0.999])

        xbins = np.linspace(xspan[0],xspan[1],200)
        ybins = np.linspace(yspan[0],yspan[1],200)
        H, xed, yed = np.histogram2d(xcolors,ycolors,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        plt.imshow(np.log10(H.T),extent=extent,aspect='auto',interpolation='nearest',origin='lower',cmap=plt.cm.Greys)
        
        plt.plot(stellar_locus[xlocusname],stellar_locus[ylocusname],color='b')
        plt.plot(stellar_locus[xlocusname],stellar_locus[ylocusname] + 2.*stellar_locus[ylocusname+"sig"],color='r',linestyle='--')
        plt.plot(stellar_locus[xlocusname],stellar_locus[ylocusname] - 2.*stellar_locus[ylocusname+"sig"],color='r',linestyle='--')
        plt.xlabel(xcolor_name)
        plt.ylabel(ycolor_name)
        plt.minorticks_on()

    plt.show()

def find_minmax(quants):
    for ii in range(len(quants)):
        print min(quants[ii]),max(quants[ii])
```

```python
def nikutta_orich_crich(w1,w2,w3,w4):
    col12 = w1-w2
    col23 = w2-w3
    col34 = w3-w4
    
    orich = np.where((col12 < 2.0) & (col12 > 0.2) & (col23 > 0.4) & (col23 < 2.2) & (col34 > 0) &
        (col34 < 1.3) & (col34 > 0.722*col23 - 0.289))

    crich = np.where((col12 > 0.629*col23 - 0.198) & (col12 < 0.629*col23 + 0.359) &
        (col34 > 0.722*col23 - 0.911) & (col34 < 0.722*col23 - 0.289))
    
    return orich[0], crich[0]
```

```python
def spectype_relation(coljk, colk3, col12, col23, col34, stype, relation):
    absw1 = np.zeros((len(stype),5))
    if relation == 'lmc':
        orich = np.array([[-8.67,-0.02],[-8.65,-0.04],[-8.39,-0.64],[-9.44,0.22],[-8.76,0.07]])
        crich = np.array([[-8.52,-0.10],[-8.98,0.03],[-8.48,-0.44],[-9.18,0.14],[-9.11,0.38]])
        
    if relation == 'bulge':
        orich = np.array([[-7.70,0.12],[-6.93,-0.33],[-7.08,-1.68],[-7.97,0.37],[-6.97,-0.74]])
        crich = np.array([[-7.60,0.08],[-7.43,0.02],[-6.92,-1.77],[-7.55,0.14],[-6.92,-0.55]])

    xvars = np.array([coljk, colk3, col12, col23, col34])
    for ii in range(len(xvars)):
        if stype[ii] == 0:
            absw1[stype == 0,ii] = orich[ii,0] + orich[ii,1]*xvars[ii,stype==0]## w1 vs colk3

        if stype[ii] == 1:
            absw1[stype == 1,ii] = crich[ii,0] + crich[ii,1]*xvars[ii,stype==1]## w1 vs colk3

    absw1 = absw1.mean(axis=1)
    return absw1
```
