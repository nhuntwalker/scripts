<b>Purpose:</b><br/>
The purpose of this notebook is to manipulate the data obtained from TRILEGAL, then analyzing how well they match up with Nikutta et al 2014's color cuts in WISE color space. The initial retrieval for the sample is as follows (this may change later):

<ul>
<li>24 files with the area of 20 square degrees each area</li>
<li> 15 < gal l < 50 </li>
<li> |gal b| < 5 </li>
</ul>

Each run had the following parameters:
<ul>
<li>M-star dust: tab_dust_dpmod60alox40_2mass_wise.dat</li>
<li>C-star dust: tab_dust_AMCSIC15_2mass_wise.dat</li>
<li>W4 < 9.0, with magnitude resolution of 0.2 mags</li>
<li>R$_\odot$: 8 kpc</li>
<li>Z$_\odot$: 0 kpc</li>
<li>IMF: Chabrier Lognormal with binary fraction of 0.3</li>
<li>No interstellar extinction (the comparison sample has presumably been dust-corrected)</li>
<li>Thin disk: $sech^2$ </li>
<li>Thick disk: $sech^2$ </li>
<li>Halo: power law</li>
<li>Bulge: Something from Binney et al. 1997</li>
</ul>

<ul>
<li>J-K > 1.0</li>
<li>W3-W4 < -0.83(W2-W3) + 3.37</li>
<li>W2-W3 > 0</li>
<li>[W1/W2/W3/W4] < [16.83/15.6/11.32/9.0]</li>
</ul>

The chemical criteria to separate between O-rich and C-rich AGB stars are as follows (cf. Nikutta et al 2014):

<b>O-rich:</b>
<ul>
<li>0 < W1-W2 < 2 (this is relaxed from what the paper cites)</li>
<li>0.4 < W2-W3 < 2.2</li>
<li>0 < W3-W4 < 1.3</li>
<li>W3-W4 > 0.722(W2-W3) - 0.289 </li>
</ul>

<b>C-rich:</b>
<ul>
<li>0.629(W2-W3) - 0.198 < W1-W2 < 0.629(W2-W3) + 0.359</li>
<li>0.722(W2-W3) - 0.911 < W3-W4 < 0.722(W2-W3) - 0.289 </li>
</ul>

```python
coeff = [0.34,0.21,0.15,0.11,0.09,0.13,0.13] # j,h,k,w1,w2,w3,w4

crich_indir = '/Users/Nick/Documents/AGBstuff/workstation/allwise_reboot/allwise_relaxed/candidates_relaxed/candidate_outs_crich/'
orich_indir = '/Users/Nick/Documents/AGBstuff/workstation/allwise_reboot/allwise_relaxed/candidates_relaxed/candidate_outs_orich/'

crich_files = [crich_indir+f for f in os.listdir(crich_indir) if f.endswith('.dat')]
orich_files = [orich_indir+f for f in os.listdir(orich_indir) if f.endswith('.dat')]

crich_df = pd.io.parsers.read_table(crich_files[0], sep='\t')
orich_df = pd.io.parsers.read_table(orich_files[0], sep='\t')

for ii in range(1,len(crich_files)):
    newdf = pd.io.parsers.read_table(crich_files[ii], sep='\t')
    crich_df = pd.concat([crich_df,newdf])

for ii in range(1,len(orich_files)):
    newdf = pd.io.parsers.read_table(orich_files[ii], sep='\t')
    orich_df = pd.concat([orich_df,newdf])

ocut =  (orich_df.w1err < 0.1) & (orich_df.w2err < 0.1) & (orich_df.w3err < 0.1) & (orich_df.w4err < 0.1) & (orich_df.w1err > 0.) & (orich_df.w2err > 0.) & (orich_df.w3err > 0.) & (orich_df.w4err > 0.)
ccut =  (crich_df.w1err < 0.1) & (crich_df.w2err < 0.1) & (crich_df.w3err < 0.1) & (crich_df.w4err < 0.1) & (crich_df.w1err > 0.) & (crich_df.w2err > 0.) & (crich_df.w3err > 0.) & (crich_df.w4err > 0.)

orich_df = orich_df[ocut]
crich_df = crich_df[ccut]


crich_df['Av'] = 3.1*crich_df.ebvs
crich_df['Ar'] = 2.751*crich_df.ebvs
crich_df['Aj'] = coeff[0]*2.751*crich_df.ebvs
crich_df['Ah'] = coeff[1]*2.751*crich_df.ebvs
crich_df['Ak'] = coeff[2]*2.751*crich_df.ebvs
crich_df['Aw1'] = coeff[3]*2.751*crich_df.ebvs
crich_df['Aw2'] = coeff[4]*2.751*crich_df.ebvs
crich_df['Aw3'] = coeff[5]*2.751*crich_df.ebvs
crich_df['Aw4'] = coeff[6]*2.751*crich_df.ebvs
crich_df['DM'] = crich_df.w1 - crich_df.absw1 - crich_df.Aw1
crich_df['dist'] = 10**((crich_df['DM']+5.)/5.)/1E3 #kpc

crich_df['coljk'] = crich_df.j - crich_df.k
crich_df['coljh'] = crich_df.j - crich_df.h
crich_df['colhk'] = crich_df.h - crich_df.k
crich_df['colk3'] = crich_df.k - crich_df.w1
crich_df['colk3'] = crich_df.k - crich_df.w3
crich_df['col12'] = crich_df.w1 - crich_df.w2
crich_df['col23'] = crich_df.w2 - crich_df.w3
crich_df['col34'] = crich_df.w3 - crich_df.w4

a = np.array(crich_df.gall)
a[np.where(a > 180)] = a[np.where(a > 180)] - 360.
crich_df.gall = pd.Series(a, index=crich_df.index)


orich_df['Av'] = 3.1*orich_df.ebvs
orich_df['Ar'] = 2.751*orich_df.ebvs
orich_df['Aj'] = coeff[0]*2.751*orich_df.ebvs
orich_df['Ah'] = coeff[1]*2.751*orich_df.ebvs
orich_df['Ak'] = coeff[2]*2.751*orich_df.ebvs
orich_df['Aw1'] = coeff[3]*2.751*orich_df.ebvs
orich_df['Aw2'] = coeff[4]*2.751*orich_df.ebvs
orich_df['Aw3'] = coeff[5]*2.751*orich_df.ebvs
orich_df['Aw4'] = coeff[6]*2.751*orich_df.ebvs
orich_df['DM'] = orich_df.w1 - orich_df.absw1 - orich_df.Aw1
orich_df['dist'] = 10**((orich_df['DM']+5.)/5.)/1E3 #kpc

orich_df['coljk'] = (orich_df.j - orich_df.k) - (orich_df.Aj - orich_df.Ak)
orich_df['coljh'] = (orich_df.j - orich_df.h) - (orich_df.Aj - orich_df.Ah)
orich_df['colhk'] = (orich_df.h - orich_df.k) - (orich_df.Ah - orich_df.Ak)
orich_df['colk3'] = (orich_df.k - orich_df.w1) - (orich_df.Ak - orich_df.Aw1)
orich_df['colk3'] = (orich_df.k - orich_df.w3) - (orich_df.Ak - orich_df.Aw3)
orich_df['col12'] = orich_df.w1 - orich_df.w2
orich_df['col23'] = orich_df.w2 - orich_df.w3
orich_df['col34'] = orich_df.w3 - orich_df.w4

a = np.array(orich_df.gall)
a[np.where(a > 180)] = a[np.where(a > 180)] - 360.
orich_df.gall = pd.Series(a, index=orich_df.index)
```

```python
def sixplots(dataframe, cut=None, title=None, scatter='no', savename=None, contours=False, nikutta_cuts=False):
    if type(cut) != type(None):
        dataframe = dataframe[cut]
        
    w1 = dataframe.w1
    w2 = dataframe.w2
    w3 = dataframe.w3
    w4 = dataframe.w4
    j = dataframe.j
    h = dataframe.h
    k = dataframe.k
    
    col12 = dataframe.col12
    col23 = dataframe.col23
    col34 = dataframe.col34
    coljh = dataframe.coljh
    colhk = dataframe.colhk
    coljk = dataframe.coljk
    colk3 = dataframe.colk3
    
    thick = 2
    tint = 0.3
    
    fig = plt.figure(figsize=(12,8.5))
    if title != None:
        fig.subplots_adjust(top=0.9,bottom=0.1,left=0.07,right=0.9, hspace=0.2)
        plt.suptitle(title)
    
    else:
        fig.subplots_adjust(top=0.975,bottom=0.1,left=0.07,right=0.9, hspace=0.2)
    ## K vs J-K
    ax = plt.subplot(2,3,1)
    xbins = np.arange(1,10.1,0.1)
    ybins = np.arange(2,15.6,0.1)
    if scatter.lower() != 'no':
        ax.scatter(coljk,k,s=1,edgecolor='none',c='k')
    else:
        H, xed, yed = np.histogram2d(coljk,k,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
        
    if contours != False:
        xspan = sp.mquantiles(np.array(coljk),[0.001,0.999])
        yspan = sp.mquantiles(np.array(k),[0.001,0.999])

        xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
        ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
        H, xed, yed = np.histogram2d(coljk,k,bins=(xbins,ybins))
        allquants = np.log10(H.ravel())
        
        levels = [0.5,1.0,1.5,2.0,2.5,3.0]
        extent=[xed[0],xed[-1],yed[0],yed[-1]]
        im = ax.contour(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', levels=levels)#, colors='r')            
        
        cax = fig.add_axes([0.925, 0.1, 0.05, 0.8]) 
        cbar = plt.colorbar(im, cax = cax)  
        cbar.ax.set_ylabel('log$_{10}$(N)')

    ax.minorticks_on()
    ax.set_xlim(1,8)
    ax.set_ylim(15.5,2)
    ax.set_xlabel('J-K (mag)')
    ax.set_ylabel('K (mag)')
    
    ## K-W3 vs J-K
    ax = plt.subplot(2,3,2)
    xbins = np.arange(1,10.1,0.1)
    ybins = np.arange(-2,10.5,0.1)
    if scatter.lower() != 'no':
        ax.scatter(coljk,colk3,s=1,edgecolor='none',c='k')
    else:
        H, xed, yed = np.histogram2d(coljk,colk3,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)

    if contours != False:
        xspan = sp.mquantiles(np.array(coljk),[0.001,0.999])
        yspan = sp.mquantiles(np.array(colk3),[0.001,0.999])

        xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
        ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
        H, xed, yed = np.histogram2d(coljk,colk3,bins=(xbins,ybins))
        allquants = np.log10(H.ravel())
        
        levels = [0.5,1.0,1.5,2.0,2.5,3.0]
        extent=[xed[0],xed[-1],yed[0],yed[-1]]
        im = ax.contour(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', levels=levels)#, colors='r')            
        
        cbar = plt.colorbar(im, cax = cax)  
        cbar.ax.set_ylabel('log$_{10}$(N)')

    ax.minorticks_on()
    ax.set_xlim(1,8)
    ax.set_ylim(0,6)
    ax.set_xlabel('J-K (mag)')
    ax.set_ylabel('K-W3 (mag)')
    
    ## K vs W2-W3
    ax = plt.subplot(2,3,3)
    xbins = np.arange(0,4.6,0.05)
    ybins = np.arange(2,15.6,0.1)
    if scatter.lower() != 'no':
        ax.scatter(col23,k,s=1,edgecolor='none',c='k')
    else:
        H, xed, yed = np.histogram2d(col23,k,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
    ax.minorticks_on()
    ax.set_xlim(0,3.5)
    ax.set_ylim(15.5, 2)
    ax.set_xlabel('W2-W3 (mag)')
    ax.set_ylabel('K (mag)')
    
    ## K-W3 vs W1-W2
    ax = plt.subplot(2,3,4)
    xbins = np.arange(-1.5,4.6,0.05)
    ybins = np.arange(-2,10.5,0.1)
    if scatter.lower() != 'no':
        ax.scatter(col12,colk3,s=1,edgecolor='none',c='k')
    else:
        H, xed, yed = np.histogram2d(col12,colk3,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
    ax.minorticks_on()
    ax.set_xlim(-0.25,2.)
    ax.set_ylim(0,6)
    ax.set_xlabel('W1-W2 (mag)')
    ax.set_ylabel('K-W3 (mag)')
    
    ## W1-W2 vs W2-W3
    ax = plt.subplot(2,3,5)
    ybins = np.arange(-1.5,4.6,0.05)
    xbins = np.arange(0,4.6,0.05)
    if scatter.lower() != 'no':
        ax.scatter(col23,col12,s=1,edgecolor='none',c='k')
    else:
        H, xed, yed = np.histogram2d(col23,col12,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)

    if nikutta_cuts == True:
        ax.fill_between([0,4], [0.629*(0) - 0.198,0.629*4 - 0.198], [0.629*(0) + 0.359,0.629*4 + 0.359], color='r', alpha=tint)
        ax.plot([0,4], [0.629*(0) + 0.359,0.629*4 + 0.359], color='r')
        ax.plot([0,4], [0.629*(0) - 0.198,0.629*4 - 0.198], color='r')
        ax.plot([0,0], [-1,4], color='k')
        ax.plot([4,4], [-1,4], color='k')
        
        ax.fill_between([0.4,2.2],[0.0,0.0],[2.0,2.0], color='b', alpha=tint)
        ax.plot([0.4,0.4], [0.0,2.0], color='b')
        ax.plot([2.2,2.2], [0.0,2.0], color='b')
    
    ax.minorticks_on()
    ax.set_xlim(0,4.)
    ax.set_ylim(0,3.)
    ax.set_xlabel('W2-W3 (mag)')
    ax.set_ylabel('W1-W2 (mag)')
    
    ## W3-W4 vs W2-W3
    ax = plt.subplot(2,3,6)
    xbins = np.arange(0,4.6,0.05)
    ybins = np.arange(-1.5,3.0,0.05)
    if scatter.lower() != 'no':
        ax.scatter(col23,col34,s=1,edgecolor='none',c='k')
    else:
        H, xed, yed = np.histogram2d(col23,col34,bins=(xbins,ybins))
        extent = [xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), interpolation='nearest', aspect="auto", extent=extent, origin='lower', cmap=plt.cm.Greys)
    
    if nikutta_cuts == True:
        ax.fill_between([0,4], [0.722*(0) - 0.911,0.722*4 - 0.911], [0.722*(0) - 0.289,0.722*4 - 0.289], color='r', alpha=tint)
        ax.plot([0,4], [0.722*(0) - 0.289,0.722*4 - 0.289], color='r')
        ax.plot([0,4], [0.722*(0) - 0.911,0.722*4 - 0.911], color='r')
        ax.plot([0,0], [-1,4], color='k')
        ax.plot([4,4], [-1,4], color='k')
        
        ax.fill_between([0.4,2.2],[0.722*0.4-0.289, 0.722*2.2-0.289],[1.3,1.3], color='b', alpha=tint)
        ax.plot([0.4,2.2], [1.3,1.3], color='b')
        ax.plot([0.4,0.4], [0,1.3], color='b')
        ax.plot([0.4,2.2], [0.722*0.4-0.289, 0.722*2.2-0.289], color='b')
        ax.plot([0,0], [-1,4], color='k')
        ax.plot([4,4], [-1,4], color='k')
    
    ax.minorticks_on()
    ax.set_xlim(0,3.5)
    ax.set_ylim(-1.,2.5)
    ax.set_xlabel('W2-W3 (mag)')
    ax.set_ylabel('W3-W4 (mag)')
    
    if savename != None:
        plt.savefig(savename)
    
    plt.show()
```

```python
def plot_photometry(dataframe,xlabels,ylabels,cut=None,histo=True,contours=False,yaxismag=False,set_xaxis=None,set_yaxis=None,cmap=plt.cm.Greys):
    if type(cut) != type(None):
        dataframe = dataframe[cut]
    
    xdata = dataframe[xlabels]
    ydata = dataframe[ylabels]
    
    fig = plt.figure(figsize=(4,4))
    ax = plt.subplot(111)
    xspan = sp.mquantiles(np.array(xdata),[0.001,0.999])
    yspan = sp.mquantiles(np.array(ydata),[0.001,0.999])

    if histo == False:
        ax.scatter(xdata,ydata,s=1,edgecolor='None',c='k')
    else:
        xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
        ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
        H, xed, yed = np.histogram2d(xdata,ydata,bins=(xbins,ybins))
        extent=[xed[0],xed[-1],yed[0],yed[-1]]
        ax.imshow(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', cmap=cmap)
    
    if contours != False:
        xbins = np.linspace(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1,101)
        ybins = np.linspace(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1,101)
        H, xed, yed = np.histogram2d(xdata,ydata,bins=(xbins,ybins))
        allquants = np.log10(H.ravel())
        #levels = np.linspace(0,max(allquants),7)[1:-1]
        levels = [0.5,1.0,1.5,2.0,2.5]
        extent=[xed[0],xed[-1],yed[0],yed[-1]]
        im = ax.contour(np.log10(H.T), extent=extent, interpolation='nearest', aspect='auto', origin='lower', colors='r',levels=levels)            
        
        cax = fig.add_axes([0.925, 0.125, 0.05, 0.775]) 
        cbar = plt.colorbar(im, cax = cax)  
        cbar.ax.set_ylabel('log$_{10}$(N)')
    
    if set_xaxis != None:
        ax.set_xlim(set_xaxis[0], set_xaxis[1])
    
    else:
        ax.set_xlim(np.round(xspan[0],1), np.round(xspan[1],1) + 0.1)

    if yaxismag == True:
        if set_yaxis != None:
            ax.set_ylim(set_yaxis[1], set_yaxis[0])
        else:
            ax.set_ylim(np.round(yspan[1],1) + 0.1,np.round(yspan[0],1))
    else:
        if set_yaxis != None:
            ax.set_ylim(set_yaxis[0], set_yaxis[1])
        else:
            ax.set_ylim(np.round(yspan[0],1), np.round(yspan[1],1) + 0.1)
            
    ax.set_xlabel(xlabels)
    ax.set_ylabel(ylabels)
    ax.minorticks_on()
                
    plt.show()
```
