What's interesting here is the differences in scale height and scale length. Scale length is going to be somewhat more difficult, as the radius distribution is clearly the product of at least two functions, AND it's incomplete toward the galactic center.

```python
#################################################
## Write function to calculate scale height for 
## a given bin
#################################################
def func(x,a,b):
    return a*np.exp(-x/b)

def func_double(x,a,b,c,d):
    return a*np.exp(-x/b) + c*np.exp(-x/d)

def scale_height(zvals, nfuncs = 2):
    dz = 0.2
    zbins = np.arange(0,20+dz,dz)
    H,ed = np.histogram(zvals, bins=zbins)
    eyeheight = [0.5,0.1]
    eyenorm = [0.3*len(zvals),1.0*len(zvals)]
    if nfuncs == 1:
        coeffs, covar = curve_fit(func, zbins[:-1], H, p0=(eyenorm[0],eyeheight[0]))
    else:
        coeffs, covar = curve_fit(func_double, zbins[:-1], H, p0=(eyenorm[0], eyeheight[0], eyenorm[1], eyeheight[1]))
    
    return coeffs
```

Note that these scale heights aren't necessarily trustworthy. Looking at the previous plots, you can see the fit for a that data isn't all that good. I need to figure out how to get model fits for a sum of exponentials.  The functional form:

$N(z|R=R_0) = \alpha\exp(-|Z|/Z_{h1}) + \beta\exp(-|Z|/Z_{h2})$

I think that this is a job for MCMC

```python
import pymc
```

```python
ii = 6

dr = 1
dz = 0.1
radbins = np.arange(0,30+dr,dr)
zbins = np.arange(0,5+dz,dz)

cut = np.where((radii[select] > radbins[ii]) &
               (radii[select] < radbins[ii+1]))
zvals = z[select][cut]
H, ed = np.histogram(abs(zvals), bins=zbins)
```

```python
## Set up MCMC sampling

dalpha = 0.5*eyenorm[0]
alph_low, alph_high = eyenorm[0]-dalpha, eyenorm[0]+dalpha
alph_rand = (alph_high - alph_low)*np.random.random(1)[0] + alph_low
alpha = pymc.Uniform('alpha', alph_low, alph_high, value=alph_rand)

dbeta = 0.5*eyenorm[1]
beta_low, beta_high = eyenorm[1]-dbeta, eyenorm[1]-dbeta
beta_rand = (beta_high - beta_low)*np.random.random(1)[0] + beta_low
beta = pymc.Uniform('beta', beta_low, beta_high, value=beta_rand)

dzh = 0.05
zh1_low = dzh
zh2_low = dzh
zh1_high = 1.
zh2_high = 3.
zh1_rand = (zh1_high - zh1_low)*np.random.random(1)[0] + zh1_low
zh2_rand = (zh2_high - zh2_low)*np.random.random(1)[0] + zh2_low
zh1 = pymc.Uniform('zh1', zh1_low, zh1_high, value=zh1_rand)
zh2 = pymc.Uniform('zh2', zh2_low, zh2_high, value=zh2_rand)

sig_low, sig_high = -2., 2.
log_sigma = pymc.Uniform('log_sigma', sig_low, sig_high, value=np.random.randint(sig_low, sig_high, 1)[0])

# Turn log_sigma into sigma
@pymc.deterministic
def sigma(log_sigma=log_sigma):
    return np.exp(log_sigma)

@pymc.deterministic
# This sets the model to be fit
def double_exp(X=zbins[:-1] + 0.5*dz, alpha=alpha, zh1=zh1, beta=beta, zh2=zh2):
    func = alpha*np.exp(-X/zh1) + beta*np.exp(-X/zh2)
    return func

# Calculate the likelihood of each model
def regress_likelihood(H, double_exp, sigma):
    func = np.sum(-np.log(sigma) + ((H - double_exp)**2)/(2.*(sigma**2)))
    return func

reg = pymc.stochastic_from_dist('reg',
                                logp = regress_likelihood,
                                dtype = np.float,
                                mv = True)

M = reg('M', double_exp, sigma, observed=True, value=H)
model = dict(H=H, alpha=alpha, zh1=zh1, beta=beta, zh2=zh2, sigma=sigma)
```

```python
# Run the MCMC sampling
S = pymc.MCMC(model)
S.sample(50000, burn=5000)
```

```python
amax = max(S.trace('alpha')[:])
bmax = max(S.trace('beta')[:])
zh1max = max(S.trace('zh1')[:])
zh2max = max(S.trace('zh2')[:])
smax = max(S.trace('sigma')[:])
```
