What's interesting here is the differences in scale height and scale length. Scale length is going to be somewhat more difficult, as the radius distribution is clearly the product of at least two functions, AND it's incomplete toward the galactic center.

```python
#################################################
## Write function to calculate scale height for 
## a given bin
#################################################
def func(x,a,b):
    return a*np.exp(-x/b)

def func_double(x,a,b,c,d):
#     return a*np.exp(-x/b) + (c/a)*np.exp(-x/d)
    return a*np.exp(-x/b) + np.exp(-x/d)

def scale_height(zvals, dz, nfuncs = 2):
    zbins = np.arange(0,5+dz,dz)
    H,ed = np.histogram(zvals, bins=zbins)
    eyeheight = [0.5,0.1]
    eyenorm = [0.3*len(zvals),1.0*len(zvals)]
    
    if nfuncs == 1:
        coeffs, covar = curve_fit(func, zbins[:-1], H, p0=(eyenorm[0],eyeheight[0]))
    else:
         coeffs, covar = curve_fit(func_double, zbins[:-1], H, p0=(eyenorm[0], eyeheight[0], eyenorm[1], eyeheight[1]))
#         coeffs, covar = curve_fit(func_double, zbins[:-1], H, p0=(eyenorm[0], eyeheight[0], 0.5*eyenorm[0], eyeheight[1]))
    
    return coeffs
```

```python
# Define the reduced chi-squared function and test the double-exponentional against curve_fit
def reduced_chi_squared(yvals, ymodel, nparams):
    N = len(yvals)
    std = np.sqrt(np.sum((yvals - ymodel)**2)/float(N))
    print 'Standard Deviation: %.3f' % std
    
    dof = len(yvals) - nparams - 1.
    print 'Degrees of Freedom: %f' % dof
    
    chi = np.sum((yvals - ymodel)**2/std**2)/dof
    return chi
```
